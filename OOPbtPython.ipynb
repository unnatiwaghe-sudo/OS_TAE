```python
#wap to find sum of two number
class Cal:
    def get(self):
        self.n1=int(input("Enter Number:"))
        self.n2 = int(input("Enter Number:"))
    def display(self):
        self.sum=self.n1+self.n2
        print(self.sum)

c1=Cal()
c1.get()   #call methode with self object
c1.display()
```
